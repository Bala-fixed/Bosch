```python
orders_df = orders_df.withColumn("total_amount", orders_df.unit_price * orders_df.quantity)
display(orders_df)
```

```python
from pyspark.sql.functions import month

orders_df = orders_df.withColumn("month", month("order_date"))
display(orders_df)
```
