```python
customers_df = spark.read.parquet("abfss://bronze@adlsteam6.dfs.core.windows.net/customers.parquet")
inventory_df = spark.read.parquet("abfss://bronze@adlsteam6.dfs.core.windows.net/inventory.parquet")
orders_df = spark.read.parquet("abfss://bronze@adlsteam6.dfs.core.windows.net/orders.parquet")
products_df = spark.read.parquet("abfss://bronze@adlsteam6.dfs.core.windows.net/products.parquet")
website_interactions_df = spark.read.parquet("abfss://bronze@adlsteam6.dfs.core.windows.net/website_interactions.parquet")
```

```python
from pyspark.sql.functions import sum as _sum

# Join orders with customers to get city info
orders_with_city_df = transformed_orders_df.join(
    customers_df,
    transformed_orders_df.customer_id == customers_df.customer_id,
    "inner"
)

# Aggregate revenue by city
city_revenue_df = orders_with_city_df.groupBy("city").agg(
    _sum("total_amount").alias("total_revenue")
)

# Get top 5 cities by revenue
top_5_cities_df = city_revenue_df.orderBy(city_revenue_df.total_revenue.desc()).limit(5)

display(top_5_cities_df)
```

```python
from pyspark.sql.functions import avg

# Group by category and calculate average stock_level
category_stock_df = products_df.groupBy("category").agg(
    avg("stock_level").alias("avg_stock_level")
)

# Get top 5 categories with lowest average stock_level
top_5_low_stock_categories_df = category_stock_df.orderBy("avg_stock_level").limit(10)

display(top_5_low_stock_categories_df)
```

```python
from pyspark.sql.functions import avg

# Join website_interactions_df with customers_df to get loyalty_status
interactions_with_loyalty_df = website_interactions_df.join(
    customers_df.select("customer_id", "loyalty_status"),
    on="customer_id",
    how="inner"
)

# Calculate average session duration for each loyalty_status
avg_session_duration_df = interactions_with_loyalty_df.groupBy("loyalty_status").agg(
    avg("time_spent_seconds").alias("avg_session_duration_seconds")
)

display(avg_session_duration_df)
```
